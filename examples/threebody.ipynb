```python
from iturtle import Screen, Turtle
import threading
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import random

def three_body_problem(t, y):
    x1, y1, x2, y2, x3, y3, vx1, vy1, vx2, vy2, vx3, vy3 = y
    dx12 = x1 - x2
    dy12 = y1 - y2
    dx13 = x1 - x3
    dy13 = y1 - y3
    dx23 = x2 - x3
    dy23 = y2 - y3
    r12 = np.sqrt(dx12**2 + dy12**2)
    r13 = np.sqrt(dx13**2 + dy13**2)
    r23 = np.sqrt(dx23**2 + dy23**2)
    F12 = -G * m1 * m2 / r12**2
    F13 = -G * m1 * m3 / r13**2
    F23 = -G * m2 * m3 / r23**2
    ax1 = F12 * dx12 / m1 + F13 * dx13 / m1
    ay1 = F12 * dy12 / m1 + F13 * dy13 / m1
    ax2 = -F12 * dx12 / m2 + F23 * dx23 / m2
    ay2 = -F12 * dy12 / m2 + F23 * dy23 / m2
    ax3 = -F13 * dx13 / m3 + -F23 * dx23 / m3
    ay3 = -F13 * dy13 / m3 + -F23 * dy23 / m3
    return [vx1, vy1, vx2, vy2, vx3, vy3, ax1, ay1, ax2, ay2, ax3, ay3]

m1, m2, m3 = 0.1, 0.1, 0.1
G = 5
x1 = random.uniform(-200, 200)
y1 = random.uniform(-200, 200)
x2 = random.uniform(-200, 200)
y2 = random.uniform(-200, 200)
x3 = random.uniform(-200, 200)
y3 = random.uniform(-200, 200)
vx1, vy1, vx2, vy2, vx3, vy3 = 0, 0, 0, 0, 0, 0
y0 = [x1, y1, x2, y2, x3, y3, vx1, vy1, vx2, vy2, vx3, vy3]
length = 400
t_span = (0, 3000)
t_eval = np.linspace(*t_span, length)

sol = solve_ivp(three_body_problem, t_span, y0, t_eval=t_eval)

def f1():
    tur1 = Turtle(screen)
    tur1.pencolor('red')
    tur1.teleport(sol.y[0][0], sol.y[1][0])
    for i in range(length):
        tur1.setheading(tur1.towards(sol.y[0][i], sol.y[1][i]))
        tur1.goto(sol.y[0][i], sol.y[1][i])

def f2():
    tur2 = Turtle(screen)
    tur2.pencolor('blue')
    tur2.teleport(sol.y[2][0], sol.y[3][0])
    for i in range(length):
        tur2.setheading(tur2.towards(sol.y[2][i], sol.y[3][i]))
        tur2.goto(sol.y[2][i], sol.y[3][i])

def f3():
    tur3 = Turtle(screen)
    tur3.teleport(sol.y[4][0], sol.y[5][0])
    for i in range(length):
        tur3.setheading(tur3.towards(sol.y[4][i], sol.y[5][i]))
        tur3.goto(sol.y[4][i], sol.y[5][i])

screen = Screen()
t1 = threading.Thread(target=f1)
t2 = threading.Thread(target=f2)
t3 = threading.Thread(target=f3)
t1.start()
t2.start()
t3.start()
t1.join()
t2.join()
t3.join()
```
