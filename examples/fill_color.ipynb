```python
from iturtle import Screen, Turtle

import time

screen = Screen()
screen.setup(1200, 800)

turtle = Turtle(screen)
turtle.color('red', 'blue')
turtle.begin_fill()
turtle.circle(150, 90)
turtle.goto(-250, 250)
turtle.left(80)
turtle.forward(100)
turtle.right(70)
turtle.forward(100)
turtle.end_fill()

turtle.done()
```

```python
screen = Screen()

turtle = Turtle(screen)

turtle.color('green', 'red')
turtle.pensize(6)

turtle.begin_fill()
for i in range(5):
    turtle.forward(120)
    turtle.right(144)
turtle.end_fill()

turtle.done()
```

```python
# 理论上执行done以后，海龟的活动线程都应该被清理
import threading

for thread in threading.enumerate():
  print(f"线程ID: {thread.ident}, 名称: {thread.name}, 是否存活: {thread.is_alive()}")
```
